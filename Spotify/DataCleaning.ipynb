```python
import pandas as pd
```

```python
df = pd.read_csv('most-streamed-artist-monthly.csv')
```

```python
df
```

```python
import re
def remove_reference(item):
    if item == 'Camilla Cabello':
        return 'Camila Cabello'
    if item == 'drake':
        return 'Drake'
    return re.sub(r'\[\d+\]','',item).strip()
```

```python
df['Artist'] = df['Artist'].map(remove_reference)
```

```python
df.set_index("Date")
```

```python
sorted(df['Artist'].unique())
```

```python
original_data_row = {}
for artist in df['Artist']:
    original_data_row[artist] = 0
```

```python
date = df['Date'].unique()[0]
```

```python
monthly_ranks = df.loc[df['Date'] == date]
for i, rank in monthly_ranks.iterrows():
    artist = rank['Artist']
    listeners = int(float(rank['Monthly Listeners (Millions)'])*1000000)
    data_row[artist] = listeners
data_row['Date'] = date
```

```python
df.loc[df['Monthly Listeners (Millions)'].isnull()]
```

```python
data = []
for date in df['Date'].unique():
    data_row = original_data_row.copy()
    monthly_ranks = df.loc[df['Date'] == date]
    for i, rank in monthly_ranks.iterrows():
        artist = rank['Artist']
        listeners = int(float(rank['Monthly Listeners (Millions)'])*1000000)
        data_row[artist] = listeners
    data_row['Date'] = date
    data.append(data_row)
```

```python
data
```

```python
new_df = pd.DataFrame.from_dict(data)
```

```python
new_df.set_index('Date')
```

```python
new_df.to_csv('most-streamed-artist-monthly-cleaned.csv', index=False)
```
